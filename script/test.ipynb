```python
# create grids based on this gdf file
thu

# get the x y range of the tsinghua campus
thu.total_bounds

# create a grid based on the total bounds of the tsinghua campus
import numpy as np
from shapely.geometry import Polygon

# create a grid based on the total bounds of the tsinghua campus
xmin, ymin, xmax, ymax = thu.total_bounds
xmin = xmin - 0.001
ymax = ymax + 0.001
length = 0.0007
wide = 0.0007
cols = list(np.arange(xmin, xmax, wide))
rows = list(np.arange(ymin, ymax, length))
rows.reverse()

polygons = []
for x in cols:
    for y in rows:
        polygons.append(Polygon([(x, y), (x+wide, y), (x+wide, y-length), (x, y-length)]))

grid = gpd.GeoDataFrame({'geometry': polygons})
grid.crs = 'epsg:4326'

# plot with black boundary
grid.plot(linewidth=0.5, edgecolor='black', alpha=0.3)
```

```python
sjoin_intersects['length'] = sjoin_intersects['geometry'].to_crs('4509').length
# sjoin_intersects.loc[sjoin_intersects.index[0]]
sjoin_intersects = sjoin_intersects[['start_time', 'end_time', 'start', 'end', 'length', 'geometry']]
```

```python
from tqdm import tqdm
for i in tqdm(grid.index):
    grid.loc[i,'keep_length'] = 0
    grid.loc[i,'length_average'] = 0
    grid.loc[i,'keep_num'] = 0
    clip = gpd.overlay(sjoin_intersects, grid[i:i+1], how='intersection').to_crs('4509')
    if len(clip)>=1:
        for j in clip.index:
            grid.loc[i,'length_average'] = grid.loc[i,'length_average'] + clip.loc[j,'length']  # 跑过的人平均的长度(总长度)
            grid.loc[i,'keep_length'] = grid.loc[i,'keep_length'] + clip.loc[j,'geometry'].length # 热力图(based ǒn length)
            grid.loc[i,'keep_num'] = grid.loc[i,'keep_num'] + 1 # 跑过的人数
    else:
        pass
```

```python
grid.plot(column='keep_length', cmap='OrRd', legend=True, linewidth=0.5,  alpha=0.3)
plt.title('The length of the trajectories in each grid')
```

```python
grid.plot(column='keep_num', cmap='OrRd', legend=True, linewidth=0.5,  alpha=0.3)
plt.title('Track frequency in each grid')
```

## 1. Encoding

**Grid-based Encoding**: This method divides the area of interest into grid cells and maps the trajectories onto these cells. Trajectories are encoded by recording sequences of grid cells they pass through. This method is straightforward but may lose precise location information.

**Vector-based Encoding**: Vector encoding uses the coordinates of trajectory points directly. This encoding retains accurate location information but can lead to very high-dimensional data for long trajectories.

**Graph-based Encoding**: Trajectories are represented as graphs, with nodes representing locations and edges representing movements. This method is especially useful for representing movements within networks, such as road or public transportation networks.

**Symbolic Encoding**: This method defines a set of symbols, each representing a feature of the trajectory (e.g., direction, speed), and then converts the trajectory into a sequence of these symbols. This approach can compress trajectory data effectively while preserving key features.

Deep Learning-based Autoencoders: Deep learning models, like autoencoders, are used to learn efficient representations of trajectory data. Autoencoders discover low-dimensional features in the data through unsupervised learning, which can be useful for dimensionality reduction and feature extraction.

Time Series-based Encoding: Utilizing time series analysis methods, such as Recurrent Neural Networks (RNNs) or Long Short-Term Memory networks (LSTMs), to directly process sequences of time and location data. This method captures the dynamic features of trajectories over time.

Attention Mechanism-based Encoding: Models based on attention mechanisms, like the Transformer, have shown excellent performance in encoding sequence data recently. This method adaptively focuses on important parts of the trajectory sequence, improving encoding efficiency and prediction accuracy.
